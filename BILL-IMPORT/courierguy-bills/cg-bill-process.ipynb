```python
import pandas as pd
from datetime import datetime

importFile = "invoices_henre-083631_04-Sep-2023-10-09.csv"
dfCG = pd.read_csv(importFile,skiprows=1)

columnsToKeep = ["Invoice no.","Invoice date","Subtotal"]
dfCG2 = dfCG[columnsToKeep]

new_column_names = {
    'Invoice date': 'Bill Date',
    'Invoice no.': 'Bill Number',
    'Subtotal': 'Rate',
    
}
dfCG3 = dfCG2.rename(columns=new_column_names)

vendorName = ['Courier Guy'] * len(dfCG3)
dfCG3["Vendor"] = vendorName
Bill_Status = ['Draft'] * len(dfCG3)
dfCG3["Bill Status"] = Bill_Status
Account = ['2340'] * len(dfCG3)
dfCG3["Account"] = Account

# Define a function to transform the date format
def transform_date(date_str):
    parsed_date = datetime.strptime(date_str, "%d %b %Y")
    formatted_date = parsed_date.strftime("%Y-%m-%d")
    return formatted_date

# Apply the function to the entire 'dates' column
dfCG3['Bill Date'] = dfCG3['Bill Date'].apply(transform_date)
```

```python
outputFile = "courier-guy-Aug-output-file.csv"
dfCG3.to_csv(outputFile, index=None)
```
